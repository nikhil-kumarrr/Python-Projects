```python
import random
import time

board = [" "]*9

def show():
    print()
    print(board[0], "|", board[1], "|", board[2])
    print("--|---|--")
    print(board[3], "|", board[4], "|", board[5])
    print("--|---|--")
    print(board[6], "|", board[7], "|", board[8])
    print()

def check_winner(player):
    lines = [
        [0,1,2],[3,4,5],[6,7,8],   
        [0,3,6],[1,4,7],[2,5,8],   
        [0,4,8],[2,4,6]            
    ]
    for line in lines:
        if board[line[0]] == board[line[1]] == board[line[2]] == player:
            return True
    return False

def user_move():
    move = input("Your move (1-9): ")
    print("You entered:", move) 

    if move.isdigit():
        position = int(move) - 1
        if 0 <= position <= 8 and board[position] == " ":
            board[position] = "X"
        else:
            print("Invalid move! Try again.")
            user_move()
    else:
        print("Please enter a number.")
        user_move()

def ai_move():
    print("AI is thinking...", end="")
    time.sleep(1)
    empty = []
    for i in range(9):
        if board[i] == " ":
            empty.append(i)
    if empty:
        choice = random.choice(empty)
        board[choice] = "O"
        print(f" AI played at position {choice + 1}")

print("Tic Tac Toe Game")
print("You are X | AI is O")
show()

while True:
    user_move()
    show()
    
    if check_winner("X"):
        print(" You won! Well played!")
        break

    if " " not in board:
        print("It's a draw!")
        break

    ai_move()
    show()

    if check_winner("O"):
        print(" AI won! Better luck next time.")
        break

    if " " not in board:
        print("It's a draw!")
        break
```
