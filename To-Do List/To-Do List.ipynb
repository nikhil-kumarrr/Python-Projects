```python
import time

tasks = []

def print_divider():
    print("----------------------")
    time.sleep(1)

def show_menu():
    print("\n To-Do List Menu:")
    print("1️  Add Task")
    print("2️  View Tasks")
    print("3️  Remove Task")
    print("4️  Exit")

def add_task():
    task = input("Enter task: ")
    print("You entered:", task)
    tasks.append(task)
    print(f'Task "{task}" added!')
    print_divider()  

def view_tasks():
    if not tasks:
        print("No tasks in the list.")
    else:
        print("\n Your Tasks:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
    print_divider()  

def remove_task():
    view_tasks()
    if tasks:
        try:
            num = int(input("Enter task number to remove: "))
            print("You entered:", num)
            if 1 <= num <= len(tasks):
                removed = tasks.pop(num - 1)
                print(f'Task "{removed}" removed.')
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")
    print_divider()  

while True:
    show_menu()
    choice = input("\nEnter your choice (1-4): ")
    print("You selected:", choice)

    if choice == '1':
        add_task()
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        remove_task()
    elif choice == '4':
        print("Exiting To-Do List.")
        break
    else:
        print("Invalid choice. Please enter 1 to 4.")
    print_divider()
```
